```python
from sklearn.datasets import make_classification

import numpy as np

X, y = make_classification(n_samples=100, n_features=2, n_informative=1, n_redundant=0, n_classes=2, n_clusters_per_class=1, random_state=41, hypercube=False, class_sep=15)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(X[:,0], X[:,1], c=y, cmap="winter", s=100)
```

```python
X[:5], y[:5]
```

```python
X.shape, y.shape
```

```python
def perceptron(X, y):
  w1 = w2 = b = 1

  lr = 0.1

  for j in range(1000):

    for i in range(X.shape[0]):

      z = w1*X[i][0] + w2*X[i][1] + b

      if z*y[i] < 0:

        w1 = w1 + lr*y[i]*X[i][0]
        w2 = w2 + lr*y[i]*X[i][1]
        b = b + lr*y[i]

  return w1, w2, b
```

```python
w1, w2, b = perceptron(X, y)
```

```python
w1, w2, b
```

```python
m = -(w1/w2)
c = -(b/w2)

print(m, c)
```

```python
x_input = np.linspace(-3, 3, 100)
y_input = m*x_input + c

plt.figure(figsize=(10, 6))
plt.plot(x_input, y_input, color="red", linewidth=3)
plt.scatter(X[:,0], X[:,1], c=y, cmap="winter", s=100)
plt.ylim(-3, 2)
```

```python
from sklearn.linear_model import Perceptron

pct = Perceptron()

```

```python
pct.fit(X, y)
```

```python
pct.coef_
```

```python
pct.intercept_
```

```python
mp = -(pct.coef_[0][0]/pct.coef_[0][1])
cp = -(pct.intercept_/pct.coef_[0][1])
```

```python
print(mp, cp)
```

```python
x_input = np.linspace(-3, 3, 100)
y_input = mp*x_input + cp

plt.figure(figsize=(10, 6))
plt.plot(x_input, y_input, color="red", linewidth=3)
plt.scatter(X[:,0], X[:,1], c=y, cmap="winter", s=100)
plt.ylim(-3, 2)
```

```python
from mlxtend.plotting import plot_decision_regions

plot_decision_regions(X, y, clf=pct, legend=2)
```
